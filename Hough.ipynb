```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/eye.jpg') # road.png is the filename
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_blur = cv.medianBlur(gray, 5)
circles = cv.HoughCircles(img_blur, cv.HOUGH_GRADIENT, 1, img.shape[0]/32, param1=200, param2=40, minRadius=5, maxRadius=100)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
       cv.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
       cv.circle(img, (i[0], i[1]), 2, (0, 0, 255), 3)

cv2_imshow(img)       
```

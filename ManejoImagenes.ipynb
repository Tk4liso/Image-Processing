```python
import cv2 as cv
from google.colab.patches import cv2_imshow    
import matplotlib.pylab as plt
#cargar y abrir imagen

#para tener acceso a drive desde código:
#from google.colab import drive
#drive.mount('/content/drive')

a=cv.imread('/content/WhatsApp Image 2022-10-25 at 16.53.02.jpeg')
cv2_imshow(a)       
```

```python
#calcular histograma
HB = cv.calcHist([a],[0],None,[256],[0,256])
HG = cv.calcHist([a],[1],None,[256],[0,256])
HR = cv.calcHist([a],[2],None,[256],[0,256])
plt.plot(HB,color='blue'), plt.plot(HG, color='green'),plt.plot(HR,color='red')
plt.xlim([0,256])
plt.show()
```

```python
#convertir a grises y ver su histograma
print('Imagen en grises: ')
gray_image = cv.cvtColor(a, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
plt.hist(a.ravel(),bins = 256, range = [0,256],color='black') 
plt.show()
```

```python
#Binarización
T,BIN= cv.threshold(gray_image, 120, 255, cv.THRESH_OTSU)
cv2_imshow(BIN)
print('valor de umbral utiizado:',T)
```

```python
#suavizado:

imBlur = cv.blur(a,(11,11))
cv2_imshow(imBlur)
```

```python
#Bordes Sobel
Sobel= cv.Sobel(src=a, ddepth=cv.CV_64F, dx=1, dy=1, ksize=5)
cv2_imshow(Sobel)
```

```python
#Convertir al modelo de color HSV
ImaHSV = cv.cvtColor(a,cv.COLOR_BGR2HSV)
cv2_imshow(ImaHSV)
```

```python
import numpy as np
#transf de  Fourier
fourier = np.fft.fft2(gray_image)
origen = np.fft.fftshift(fourier) 
espectro= 50*np.log(np.abs(origen))
#plt.imshow(espectro)
plt.imshow(espectro, cmap = 'gray')
```

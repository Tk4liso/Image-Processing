```python
from sklearn.neural_network import MLPClassifier

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#cargar conjunto de entrenamiento
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/glass.csv')
data.head()

#dividir atributos y clases
datos=data[['At1', 'At2', 'At3']]
clases=data['class']
#crear conjuntos para validar resultados
X_train, X_test, y_train, y_test = train_test_split(datos, clases, test_size=0.1, random_state=50)

#entrenar red neuronal /construir modelo  con el conjunto de entrenamiento
NN = MLPClassifier(hidden_layer_sizes=(5,3,6,),verbose=True,learning_rate_init=0.1,momentum=0.05,max_iter=100, activation='tanh',);
NN.fit(X_train,y_train)

#evaluar con conjunto de prueba
respuestas=NN.predict(X_test)
# calcular la exactitud
print('Exactitud:')
accuracy_score(y_test,respuestas)
```

```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow  

Im=cv.imread('/content/drive/MyDrive/Colab Notebooks/FPN.bmp')
masc = np.ones((3,3))
cv2_imshow(Im)
op = cv.morphologyEx(Im, cv.MORPH_OPEN, masc)
clos=cv.morphologyEx(op, cv.MORPH_CLOSE, masc)
print('Opening+Closing;')
cv2_imshow(clos)
```

```python
print('Erosionada:')
er1=cv.morphologyEx(Im, cv.MORPH_ERODE, masc)
cv2_imshow(er1);
print('erosion + dilatacion  = apertura')
dil1=cv.morphologyEx(er1, cv.MORPH_DILATE, masc)
cv2_imshow(dil1)
print('Dilatacion:')
dil2=cv.morphologyEx(dil1, cv.MORPH_DILATE, masc)
cv2_imshow(dil2)
print('Dilata +erosiona = cierre')
er2=cv.morphologyEx(dil2, cv.MORPH_ERODE, masc)
cv2_imshow(er2)
```
